# Summary of the dataset
This dataset gives a detailed description of the Customer Churn at TELCO. Telco systems is telecommunications company dealing
internet, television streaming, phone etc. This means they provide an overall communication package to the customers.

Objective
Our aim is to 
1> understand the reason(s) why customers chose to leave Telco
2> if they stay, why do they do so
3> are all groups of consumers updated to the latest products and is that a factor for leaving the brand

Since telecommunication is evolving, we need to analyse if Telco is actually ensuring the latest products to consumers,
which may affect the churn rate.
